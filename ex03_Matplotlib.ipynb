```python
import matplotlib.pyplot as plt
```

### 선 그래프

```python
# plot 함수 - 일직선으로 되어있는 Lineplot 그리기

y = [2,4,6]
plt.plot(y) # 그래프 그리기
plt.show() # 그래프 출력
```

```python
# x축 설정하여 그래프 그리기
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.show()
```

```python
# line style
# 선 스타일 지정
# 변수명.plot(x,y, linestyle="--")
```

```python
x = [2,4,6,8]
y = [10,13,16,20]

y2 = [10,20,16,13]

plt.plot(x,y)
plt.plot(x,y2, ls = ':')
plt.show()
```

```python
plt.plot(x, y, color = 'r')
plt.show()
```

```python
plt.plot(x, y, color = '#9370DB')
plt.show()
```

```python
# Marker color
# marker facecolor
plt.plot(x,y,color = '#9370DB', marker = 'o', mfc = 'r')
plt.show()
```

```python
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

plt.plot(x,y, marker = 'o', mfc = 'green', ls = ':')
plt.show()
plt.plot(x,y, color = 'red', marker = 's', mfc = 'blue', ls = ':')
plt.show()
```
